# Lets build a model on the top of above dataset

```python
df.head()
```

```python
## Drop the Phone Name column
df.drop("Phone Name", axis= 1, inplace= True)
```

```python
X = df.drop("Price in INR", axis=1)
y = df["Price in INR"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

#### First standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
SS = StandardScaler()
```

```python
X_train= SS.fit_transform(X_train)
X_test=SS.transform(X_test)
```

```python
X_train
```

```python
X_test
```

###  --> Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
Model= LinearRegression()
```

```python
Model.fit(X_train, y_train)
```

```python
test_prediction = Model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
MAE = mean_absolute_error(y_test, test_prediction).round(3)
MSE = mean_squared_error(y_test, test_prediction).round(3)
RMSE = np.sqrt(mean_squared_error(y_test, test_prediction)).round(3)
print(f"MAE: {MAE}, MSE: {MSE}, RMSE: {RMSE}")
```

```python
sns.kdeplot(data=test_prediction, label="Price Predicted")
sns.kdeplot(data=y_test, label="Actual Price")
plt.legend();
```
